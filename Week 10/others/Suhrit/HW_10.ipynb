## Compute Disparity Map

```python
a, b = left.shape 
disparity = np.zeros((a, b), np.uint8)
k=50
maxsearch=50
window_size=5
pos_adjust = 255 / maxsearch

for i in range(window_size, a - window_size):      
    for j in range(window_size, b - window_size):
        best_pos = 0
        ncc1 = float("-inf")                    #initialize with lowest possible value
        #calculate 50 diff ncc for each patch 
        for pos in range(maxsearch):   

            ncc = 0

            window_left = left[i-window_size:i+window_size, j-window_size:j+window_size]
            window_right = right[i-window_size:i+window_size, j-window_size:j+window_size]

            window_left_avg = np.mean(window_left)
            window_left_std = np.std(window_left)
            window_right_avg = np.mean(window_right)
            window_right_std = np.std(window_right)

            for k in range(-window_size, window_size):
                for l in range(-window_size, window_size):
                    temp = (int(left[i+k, j+l]) - window_left_avg) * (int(right[i+k, (j+l) - pos]) - window_right_avg)
                    if window_left_std ==0 or window_right_std ==0:
                        temp = 0
                    else:
                        temp /= window_left_std*window_right_std
                    ncc += temp             
            if ncc > ncc1:
                ncc1 = ncc
                best_pos = pos

        disparity[i, j] = best_pos                         #xl-xr -> best offset
```

```python
# Display Disparity
disparity1=disparity
skimage.io.imshow(disparity1,cmap='gray')
```

## Clipping the output between 0-50

```python
a,b = disparity1.shape
ans3 = np.zeros((a,b),np.uint8)

for i in range(a):
    for j in range(b):
        if(disparity1[i][j] > 50):
            ans3[i][j] = 50
        else:
            ans3[i][j] = disparity1[i][j]
```

## Question 2

```python
train=np.loadtxt('train.txt')
print(train[0])
```

```python
#Load Training Data
xtr=train[:,0]
ytr=train[:,1]
labels=train[:,2]
print(labels.shape)
```

```python
#Split training data into classes
class1=[]
class2=[]
for i in range(1000):
    if(labels[i]==1):
        class1.append([xtr[i],ytr[i]])
    else:
        class2.append([xtr[i],ytr[i]])
class1=np.array(class1)
class2=np.array(class2)
print(class1.shape)
```

```python
#Load test data
test=np.loadtxt('test.txt')
xts=test[:,0]
yts=test[:,1]
true_labels=test[:,2]
```

```python
def compute_distances(x,y):
    X=x*np.ones(1000)
    Y=y*np.ones(1000)
    dx=X-xtr
    dy=Y-ytr
    dist=[]
    for i  in range(1000):
        dist.append(math.sqrt(dx[i]**2+dy[i]**2))
    dist=np.array(dist)
    return dist
```

```python
DistAll=[]
for i in range(1000):
    dist=compute_distances(xts[i],yts[i])
    DistAll.append(dist)
```

```python
print(len(DistAll))
print(DistAll[0].shape)
```

```python
def compute_knn(k,dist):
    a=np.argsort(dist)
    knn=labels[a[0:k]]
    #find unique values in array along with their counts
    vals, counts = np.unique(knn, return_counts=True)
    #find mode
    mode_value = np.argwhere(counts == np.max(counts))
    res=vals[mode_value].flatten().tolist()
    return res[0]
```

## Calculate KNN for K=1

```python
#Store of classes of K Nearest Neighbors
KNNClasses1=[]
k=1
for i in range(1000):
    knn_class=compute_knn(k,DistAll[i])
    KNNClasses1.append(knn_class)
```

```python
KNNClasses1[2]
```

```python
#Split test data into different classes
Class1=[]
Class2=[]
for i in range(1000):
    if(KNNClasses1[i]==1):
        Class1.append([xts[i],yts[i]])
    else:
        Class2.append([xts[i],yts[i]])
Class1=np.array(Class1)
Class2=np.array(Class2)
print(Class1.shape)
```

```python
#Calculate test data points that are misclassified and split them into their 
mislabeled=[]
for i in range(1000):
    if(KNNClasses1[i]!=true_labels[i]):
        mislabeled.append([xts[i],yts[i]])
mislabeled=np.array(mislabeled)
```

```python
#Plot the test data predicted labels and true labels
#plt.plot(class1[:,0],class1[:,1],'.',color='red')
#plt.plot(class2[:,0],class2[:,1],'.',color='blue')
print('accuracy for KNN=1', 100-mislabeled.shape[0]/10)
plt.plot(Class1[:,0],Class1[:,1],'o',color='green')
plt.plot(Class2[:,0],Class2[:,1],'o',color='blue')
plt.plot(mislabeled[:,0],mislabeled[:,1],'ko',color='red')
plt.legend(['Class 1 test','Class 2 test','Mislabelled points'],loc='lower left')
```
