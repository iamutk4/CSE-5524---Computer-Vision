### Write a function to compute 2D Gaussian derivative masks Gx and Gy with standard deviation (sigmaD) of 0.7

```python
def gaussDeriv2D(sigmaD):
    x=np.arange(-math.ceil(3*sigmaD),math.ceil(3*sigmaD),1)    
    y=np.arange(-math.ceil(3*sigmaD),math.ceil(3*sigmaD),1)    
    factor = 1/(2*math.pi*math.pow(sigmaD,4))  
    Gx=np.zeros((y.size,x.size))
    Gy=np.zeros((y.size,x.size))
    for j in range(0,y.size,1):          
        for i in range(0,x.size,1):
            Gx[j][i]=x[i]*np.exp(-(x[i]*x[i]+y[j]*y[j])/(2*sigmaD**2))*factor
    Gx = Gx / (np.sum(np.abs(Gx)))
    for j in range(0,y.size,1):          
        for i in range(0,x.size,1):      
            Gy[j][i]=y[j]*np.exp(-(x[i]*x[i]+y[j]*y[j])/(2*sigmaD**2))*factor
    Gy = Gy / (np.sum(np.abs(Gy)))
    return [Gx, Gy]

[Gx,Gy]=gaussDeriv2D(0.7)
print(np.sum(np.abs(Gx)))
print(np.sum(np.abs(Gy)))
```

### Apply gaussian derivate to get Image derivatives (Ix, Iy)

```python
# Calulcate Ix

Ix=scipy.ndimage.correlate(checkerImage, Gx, mode='nearest')

# Calculate Iy

Iy=scipy.ndimage.correlate(checkerImage, Gy, mode='nearest')

plt.subplot(1,1,1)
plt.imshow(Ix, cmap='gray')
plt.show()

plt.subplot(1,1,1)
plt.imshow(Iy, cmap='gray')
plt.show()
```

### Multiply derivates to get Ix^2, Iy^2, and IxIy

```python
IxSquare = np.square(Ix,dtype='long')
IySquare = np.square(Iy,dtype='long')
IxIy = np.multiply(Ix,Iy,dtype='long')

plt.subplot(1,1,1)
plt.imshow(IxSquare, cmap='gray')
plt.show()

plt.subplot(1,1,1)
plt.imshow(IySquare, cmap='gray')
plt.show()

plt.subplot(1,1,1)
plt.imshow(IxIy, cmap='gray')
plt.show()
```

### Apply gaussian blur with standard deviation (sigmaI) of 1

```python
sigmaI = 1
gaussianBlurIxSquare = cv.GaussianBlur(np.float32(IxSquare),(2*math.ceil(3*sigmaI)+1,2*math.ceil(3*sigmaI)+1),sigmaI)
gaussianBlurIySquare = cv.GaussianBlur(np.float32(IySquare),(2*math.ceil(3*sigmaI)+1,2*math.ceil(3*sigmaI)+1),sigmaI)
gaussianBlurIxIy = cv.GaussianBlur(np.float32(IxIy),(2*math.ceil(3*sigmaI)+1,2*math.ceil(3*sigmaI)+1),sigmaI)

plt.subplot(1,1,1)
plt.imshow(gaussianBlurIxSquare, cmap='gray')
plt.show()

plt.subplot(1,1,1)
plt.imshow(gaussianBlurIySquare, cmap='gray')
plt.show()

plt.subplot(1,1,1)
plt.imshow(gaussianBlurIxIy, cmap='gray')
plt.show()
```

### Computer Cornerness function R with trace weighting factor (alpha) = 0.05

```python
alpha = 0.05
R = (gaussianBlurIxSquare * gaussianBlurIySquare) - (gaussianBlurIxIy) ** 2 - (alpha * (gaussianBlurIxSquare + gaussianBlurIySquare) ** 2)                                                        
                                                                               
```

### Display R and values of R(17:23, 17:23)

```python
plt.subplot(1,1,1)
plt.imshow(R, cmap='gray')
plt.show()

print(R[16:22,16:22])
```

### Display thresholded R and thresholded values of R(17:23, 17:23)

```python
threshold = 1000000
thresholdedR = np.where(R >= threshold, R, 0)
plt.subplot(1,1,1)
plt.imshow(thresholdedR, cmap='gray')
plt.show()

print(thresholdedR[16:22,16:22])
```

### Perform non-maximum suppression on R

```python
thresholdedSuppressedR = thresholdedR
a,b=thresholdedSuppressedR.shape

for i in range(0,a-2,3):
    for j in range(0,b-2,3):
        maxValue = np.max(thresholdedSuppressedR[i:i+3,j:j+3])
        count = 0
        for k in range(i,i+3):
            for l in range(j,j+3):
                if(thresholdedSuppressedR[k][l] < maxValue):
                    thresholdedSuppressedR[k][l] = 0
                else:
                    count += 1
                    if(count > 1):
                        thresholdedSuppressedR[k][l] = 0
io.imshow(thresholdedSuppressedR,cmap='gray')
```

```python
image = checkerImage
image = np.where(thresholdedSuppressedR,255,image)
io.imshow(image)
```

### Discussion

I used gaussian derivative with sigma = 0.7 and gaussian smoothing with sigma = 1 to generate cornerness function R with trace weighting factor as 0.05. All the values less than 1,000,000 were removed by threshholding and non-maximum suppression was performed. Through this, we can extract corners and infer features of an image.

### Write a function to store intensities of 16 neighbor pixel locations on circular border of radius 3 for all pixels

```python
def circularNeighborIntensities(img):
    a,b = img.shape
    circularPixelIntensities = np.zeros((a,b,16))
    
    for i in range(3,a-3):
        for j in range(3,b-3):
            circularPixelIntensities[i][j] = img[i-3,j],img[i-3,j+1],img[i-2,j+2],img[i-1,j+3],img[i,j+3],img[i+1,j+3],img[i+2,j+2],img[i+3,j+1],img[i+3,j],img[i+3,j-1],img[i+2,j-2],img[i+1,j-3],img[i,j-3],img[i-1,j-3],img[i-2,j-2],img[i-3,j-1]
    
    return circularPixelIntensities
```

### Call function and store 16 intensity values for all possible circular border pixels

```python
neighborIntensities = circularNeighborIntensities(towerImage)
```

### Create Contiguous list for threshhold T = 10

```python
T = 10
a,b = towerImage.shape
contiguousList=np.zeros((a,b,16))
for i in range(3,a-3):
    for j in range(3,b-3):
        for k in range(16):
            if(neighborIntensities[i][j][k] > towerImage[i][j] + T):
                contiguousList[i][j][k] = 1
            elif(neighborIntensities[i][j][k] < towerImage[i][j] - T):
                contiguousList[i][j][k] = -1
```

### Detect FAST feature points for threshhold T = 10

```python
fast = np.zeros((a,b))
fastX = []
fastY = []
nStar = 9

for i in range(3,a-3):
    for j in range(3,b-3):
        n=0
        count=[]
        for k in range(1,16):
            if(contiguousList[i][j][k-1] != 0 and contiguousList[i][j][k] == contiguousList[i][j][k-1]):
                n += 1
            else:
                count.append(n + 1)
                n = 0
        if(len(count) == 0 and n == 15):
            fastX.append(j)
            fastY.append(i)
            continue
        if(contiguousList[i][j][0] == contiguousList[i][j][15]):
            count.append(count[0] + count[len(count)-1])
        for c in range(len(count)):
            if(count[c] > nStar):
                fastX.append(j)
                fastY.append(i)
```

```python
plt.imshow(towerImage,cmap='gray')
plt.scatter(fastX,fastY,color = 'red',s=5)
```
